```python
import pandas as pd

# Load the data
df = pd.read_csv('top_100_saas_companies_2025.csv')

#Convert G2 Rating to numeric (in case it's stored as string)
df['G2 Rating'] = pd.to_numeric(df['G2 Rating'], errors='coerce')

#Create Rating_Tier column using conditional logic (CASE WHEN style)
def classify_rating(rating):
    if pd.isna(rating):
        return 'Not Rated'
    elif rating >= 4.7:
        return 'Excellent'
    elif rating >= 4.3:
        return 'Very Good'
    elif rating >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Rating_Tier'] = df['G2 Rating'].apply(classify_rating)

#Display sample output
print("\nClassified G2 Ratings:")
print(df[['Company Name', 'G2 Rating', 'Rating_Tier']].head(20))
```
