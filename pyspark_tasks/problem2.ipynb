```python
#Import libraries
import pandas as pd

#Load CSV
df = pd.read_csv('top_100_saas_companies_2025.csv')

#Convert 'Valuation' from string to numeric value
def parse_valuation(val):
    if isinstance(val, str):
        val = val.replace('$', '').replace('B', '').strip()
        try:
            return float(val) * 1e9  # Convert billions to number
        except:
            return None
    return None

df['Valuation_Num'] = df['Valuation'].apply(parse_valuation)

# Drop rows with missing 'Industry' or 'Valuation_Num'
df = df.dropna(subset=['Industry', 'Valuation_Num'])

#Rank companies within each industry by Valuation_Num (descending)
df['Rank'] = df.groupby('Industry')['Valuation_Num'].rank(method='dense', ascending=False)

#Filter top 2 companies per industry
top_2_per_industry = df[df['Rank'] <= 2].sort_values(by=['Industry', 'Rank'])

#Display result
print("\nTop 2 companies by valuation within each industry:")
print(top_2_per_industry[['Company Name', 'Industry', 'Valuation', 'Valuation_Num', 'Rank']])
```
